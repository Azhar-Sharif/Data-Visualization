#### Import all necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Data Loading and Inspection

```python
# Load The Iris dataset
df = sns.load_dataset("iris")

# Printing First Five rows
df.head()

```

```python
# Printing Last Five rows
print(df.tail())
```

```python
# Checking Details
df.describe()
```

```python
# Checking For Any missing value
df.info()
```

```python
# Displaying The datatype of each column
df.dtypes
```

### Data Manipulation

```python
# Adding a sepal_area column in dataset
df["sepal_area"] = df["sepal_length"] * df["sepal_width"]
df
```

```python
# Extracting Only Setosa Species Rows
new_df = df[df["species"]=="setosa"]
new_df.head()
```

```python
# SOrting by petal length
df["petal_length"] = df["petal_length"]
sorted_df = df.sort_values("petal_length",ascending=False)
sorted_df
```

```python
# New dataset with a petal width greater than 1.5 cm.
new_df_petal_lengtht = df[df["petal_width"] > 1.5]
new_df_petal_lengtht.info()
# There are 56 entries With Petal Length Greater than 1.5
```

```python
# Calculating Mean for each Species
mean_sepal_width = df.groupby("species")["sepal_width"].mean()
mean_sepal_width
```

## Data Visualization

```python
# Plotting Histogram for Sepal Length
plt.hist(df["sepal_length"],bins=10,color='skyblue', edgecolor='black')

# Add title and labels for x and y axes
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Creating scatter plot Using seaborn
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')

# Adding title and labels for x and y axes
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')


plt.show()
```

```python
# Creating boxplot to show Distribution of Petal_length across Each Specie
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='petal_length')

# Adding title and labels for x and y axes
plt.title('Distribution of Petal Length for Each Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')

plt.show()
```

```python
# Creating a pairPlot of Numeric columns With species as Colour
sns.pairplot(data=df, hue='species')


plt.show()
```

## Final Observations
Sepal:
* Setosa generally has smaller sepal lengths and widths compared to the other two species.
* Versicolor and Virginica show considerable overlap in sepal lengths and widths

Petal:
* There is a clear distinction between the petal lengths and widths of Iris-setosa compared to the other two species. Iris-setosa typically has shorter and narrower petals.

There seems to be a strong positive correlation between petal length and petal width for all species, with longer petals generally corresponding to wider petals.
